```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

```python
df =  pd.read_csv('creditcard.csv')
df
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
df
```

The 'Time' feature is about how much time the given transaction take and the features from 'V1' to 'V28' are all representing an information about user's credit card.
The feature 'Amount' is telling us the amount on which user done transaction in dollars. And the last column contains binary values '0' is representing Legitimate or Normal transactions and '1' is representing Fradualent transactions.

```python
df['Class'].value_counts()
```

```python
normal = df[df.Class == 0]
normal
```

```python
fraud =  df[df.Class == 1]
fraud
```

```python
normal['Amount'].describe()
```

```python
fraud.Amount.describe()
```

```python
df.groupby('Class').mean()
```

```python
# Getting sample data from normal transactions for comparing with fraud transactions.
n_smple = normal.sample(n=156)
n_smple
```

```python
data = pd.concat([n_smple, fraud], axis=0)
data
```

```python
data['Class'].value_counts()
```

```python
data.groupby('Class').mean()
```

**Getting Model Ready**

```python
x = data.drop('Class', axis = 1)

y = data['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify=y, random_state = 2)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

**Model Evaluation**

```python
# accuracy on training data.
prediction = model.predict(x_train)
score = accuracy_score(prediction, y_train)
print(f'Training Data Prediction Accuracy: {score}')
```

```python
# accuracy on test data.
prediction = model.predict(x_test)
score = accuracy_score(prediction, y_test)
print(f'Test Data Prediction Accuracy: {score}')
```

I get the accuracy score on training data and testing data separately because by comparing both scores we can know our model is underfitted or overfitted.

underfitting" occurs when a model is too simple to capture the underlying patterns in the data, resulting in poor performance on both training and testing data, while "overfitting" happens when a model learns the training data too closely, performing well on the training data but poorly on new data due to excessive complexity, essentially "memorizing" the training data instead of generalizing to unseen examples

In our case, there is a little difference between both predictions accuracy score. So, we can say our model is good.

```python
df[['price','room_type','availability_365']]=df[['price','room_type','availability_365']].fillna('Unknown')
df[['minimum_nights','number_of_reviews','calculated_host_listings_count']]=df[['minimum_nights','number_of_reviews','calculated_host_listings_count']].fillna('Not Available')
```
